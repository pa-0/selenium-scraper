```python
import csv
from h11 import Data
import pandas as pd
import glob
import os
import numpy as np
maketrans = str.maketrans

USD_EUR_RATE = 0.99
USD_UAH_RATE = 0.034
```

```python
class DataPrep():

    def prepData(self):    
        data = pd.read_csv("Results/UA/atbmarket.com - meat.csv")
        data['amount'] = data['name'].str.split(',')
        data['amount'] = data['name']


        for index, row in data.iterrows():
            if row['price_unit'] == 'грн  /шт':
                # data.at[index, 'c1'] = 'amount'
                row['price_base'] = 0

        # split column by comma
        data['amount'] = data['name'].str.split(',').str[1]
        data['amount'] = data['amount'].astype(str).str.replace(' ', '')
        data['amount'] = data['amount'].astype(str).str.replace('к', '000')
        data['amount'] = data['amount'].astype(str).str.replace('г', '')
        
        data = data[data.amount.apply(lambda x: x.isnumeric())]
        data['amount'] = data['amount'].astype(int) / 1000
        data['price_base'] = data['price_usd'].astype(float) / data['amount'].astype(float)
        # data['price'] = data['price'].astype(str).str.replace(' ', '')
        # data['price'] = data['price'].astype(str).str.replace('€', '')
        # data['price'] = data['price'].astype(str).str.replace(',', 'COMMA')
        # data['price'] = data['price'].astype(str).str.replace('.', '')
        # data['price'] = data['price'].astype(str).str.replace('COMMA', '.')
        # data['price'] = data['price'].astype(str).str.replace('-', '0')
        # data['price'] = data['price'].astype(str).str.replace('–', '0')
        # data = data[data['price'].astype(str) != 'undefined']
        # data['price_usd'] = data['price'].astype(float) * USD_EUR_RATE
        data.to_csv("Results/UA/atbmarket.com - meat - u.csv")

 

    def prepDataManual(self):    
        filename = "Results/DE/amazon.de - iphone.csv"
        data = pd.read_csv(filename)
        data['price'] = data['price'].astype(str).str.replace(',', '')
        data['price_usd'] = data['price'].astype(float) * USD_EUR_RATE
        data = data[data['price'].astype(float) > 100]
        data.to_csv(filename.split('.csv')[0] + "- usd.csv")
        print("wrote to: ",filename.split('.csv')[0] + "- usd.csv")

    def prepDataAuto(self):    
        path ="Results/all/"
        filenames = glob.glob(path + "/*.csv")
        finaldata = []
        for index,filename in enumerate(filenames): 
            print("reading from: ",filename)
            temp = pd.read_csv(filename)
            # temp['price_usd'] = temp['price'].astype(float) * USD_EUR_RATE
            # temp['catagory'] = filename.split('.csv')[0].split('-')[-1]
            temp['country'] = filename.split('.csv')[0].split('_')[-1]
            finaldata.append(temp)
            # print(temp)
            # temp.to_csv(filename.split('.csv')[0] + " - usd.csv")
            # temp.to_csv(filename.split('.')[0] + "-usd.csv")
        finaldata = pd.concat(finaldata,sort=False,ignore_index=True)
        finaldata.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
        finaldata.to_csv("Results/all.csv")
    def dataModify(self):
        data = pd.read_csv("Results/all.csv")
        data.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
        data.to_csv("Results/all.csv")
    

prep = DataPrep()
prep.prepData()
# prep.prepDataAuto()
# prep.dataModify()
```
